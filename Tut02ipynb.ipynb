```python
import pandas as pd

data_frame = pd.read_csv("/content/drive/MyDrive/Mall_Customers.csv")
data_frame.rename({'CustomerID': "ID", 'Gender': "Sex", 'Annual Income (k$)': "Salary"}, axis=1, inplace=True)
data_frame.head()
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
label_encoder = preprocessing.LabelEncoder()
data_frame['Sex']= label_encoder.fit_transform(data_frame['Sex'])
data_frame.head(100)
```

```python
data_frame.rename({' Spending Score (1-100)':"Spending_Score"}, axis=1,
inplace=True)
data_frame.head(20)
```

```python
def satisfaction(value):
 if value > 70:
  return "Satisfied"
 else:
  return "Unsatisfied"
```

```python
print(data_frame.columns)
```

```python
data_frame['Spending Score (1-100)'].apply(satisfaction)
```

```python
data_frame['Customer_Satisfaction'] = data_frame['Spending Score (1-100)'].apply(satisfaction)
data_frame.head(20)
```

```python
data_frame.to_csv('/content/drive/MyDrive/Prepared_Mall_Customers.csv', index=False)
```
